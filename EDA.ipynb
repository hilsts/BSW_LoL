```python
def create_df_by_league(league):
    files = os.listdir('data/raw/')
    league_files = [file for file in files if league in file]
    league_data, league_entries = [], []
    
    for file in league_files:
        f = open(f'data/raw/{file}', 'r')
        league_data += json.loads(f.read())

    for i in league_data:
        league_entries += i['entries']
    
    return pd.DataFrame(league_entries)
    
```

```python
def get_events_from_match(match_id):

    frames = matches_df[matches_df['gameId'] == int(match_id)]['timeline'].tolist()[0]['frames']
#     print(frames)
    
    frame_events = [x['events'] for x in frames]
#     print(f'frame events {len(frame_events)}')
    match_events = []

    for frame_event in frame_events:
        for event in frame_event:
            match_events.append(event)
    match_events
#     print(f'match events {len(match_events)}')

    match_events_df_ = pd.DataFrame(match_events)
    match_events_df_['gameId'] = match_id
    
    return match_events_df_


def create_teams_df(matches_df):
    
    teams_records = matches_df[['gameId', 'teams']].to_dict('records')
    teams_df = pd.json_normalize(teams_records, record_path='teams', meta='gameId')

    return teams_df

# def has_dragon_soul(row):
    
    
```
